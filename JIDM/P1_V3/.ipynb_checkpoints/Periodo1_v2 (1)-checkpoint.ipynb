```python
# ==============================================================================
# 1. CONFIGURA√á√ÉO DE FILTROS (SPAM, TRAVAS E STOPWORDS)
# ==============================================================================

# Stopwords complementares (mantidas do seu c√≥digo)
custom_stopwords = {
    "bom", "dia", "boa", "tarde", "noite", "ol√°", "oi", "ola", "tchau",
    "kk", "kkk", "kkkk", "kkkkk", "rs", "rsrs", "aff", "eita", "oxi", "oxe",
    "imagem", "figura", "sticker", "√°udio", "audio", "v√≠deo", "video",
    "mensagem", "encaminhada", "apagou"
}
# Supondo que stop_words_pt j√° venha do NLTK ou spacy
# stop_words_total = stop_words_pt | custom_stopwords 
stop_words_total = custom_stopwords # (Use essa linha se n√£o tiver a lista pt carregada ainda)

punctuation = set(string.punctuation)

# Lista de termos proibidos (SPAM COMERCIAL / FINANCEIRO)
spam_keywords = [
    "limpar nome", "aumentar score", "cnpj", "cr√©dito", "empr√©stimo", 
    "venda", "oferta", "desconto", "turismo", "excurs√£o", "sa√≠da", 
    "ganhe", "oportunidade", "renda extra", "marketing", "cons√≥rcio",
    "delivery", "entregamos", "promo√ß√£o"
]

# Termos de Vandalismo / Travazap
crash_keywords = [
    "trava", "crash", "binario", "destrava", "imune", "rajada", 
    "virus", "clonagem", "hacker", "phoenix", "revaneio"
]

# ==============================================================================
# 2. FUN√á√ïES AUXILIARES
# ==============================================================================

def is_junk_message(text_lower):
    """
    Retorna True se a mensagem for considerada Lixo (Spam, Trava ou Vandalismo)
    """
    # 1. Checagem de Travazap por tamanho excessivo
    # Mensagens leg√≠timas de WhatsApp raramente passam de 3000 caracteres
    if len(text_lower) > 3000:
        return True
        
    # 2. Checagem por palavras-chave de Vandalismo
    if any(term in text_lower for term in crash_keywords):
        return True
        
    # 3. Checagem por SPAM Comercial
    if any(term in text_lower for term in spam_keywords):
        return True
        
    return False

def simplify_urls(text):
    """
    Substitui URLs completas apenas pelo dom√≠nio principal.
    Ex: 'https://terrabrasilnoticias.com/2022/noticia' -> 'terrabrasilnoticias.com'
    """
    # Regex para encontrar URLs (http, https ou www)
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    
    def extract_domain(match):
        url = match.group(0)
        # Adiciona http se faltar para o urlparse funcionar
        if not url.startswith('http'):
            url = 'http://' + url
        try:
            parsed = urlparse(url)
            domain = parsed.netloc
            # Remove o 'www.' se existir para limpar mais
            return domain.replace('www.', '')
        except:
            return url # Se falhar, devolve o original

    return url_pattern.sub(extract_domain, text)
```

```python
# 3. FUN√á√ÉO PRINCIPAL DE PRE-PROCESSAMENTO
# ==============================================================================

def preprocess_text_advanced(text):
    """
    Pipeline completo: 
    Filtra Lixo -> Normaliza URL -> Lowercase -> Demojize -> Tokenize -> Limpa
    """
    if not isinstance(text, str):
        return []

    # 1. Lowercase inicial (essencial para os filtros funcionarem)
    text = text.lower()
    
    # 2. FILTRAGEM PESADA (Se for spam/trava, retorna lista vazia imediatamente)
    if is_junk_message(text):
        return [] # Retorna vazio, indicando que essa mensagem deve ser ignorada
    
    # 3. Normaliza√ß√£o de URLs (Antes de remover pontua√ß√£o)
    text = simplify_urls(text)
    
    # 4. Normalizar emojis (transforma üòÉ em :smiley:)
    text = emoji.demojize(text)
    
    # 5. Tokeniza√ß√£o Universal
    tokens = nltk.word_tokenize(text)
    
    tokens_clean = []
    for token in tokens:
        # Remover pontua√ß√£o simples (mas cuidado, dominios tem ponto ex: google.com)
        # Se o token tiver ponto no meio (google.com), mantemos. Se for s√≥ pontua√ß√£o, removemos.
        if token in punctuation:
            continue
            
        # Remover stopwords
        if token in stop_words_total:
            continue
            
        # Regra extra: remover tokens muito curtos que n√£o sejam n√∫meros
        # (evita sujeira como 'a', 'b', 'vc')
        if len(token) < 2 and not token.isdigit():
            continue

        tokens_clean.append(token)
    
    return tokens_clean

# ==============================================================================
# EXEMPLO DE USO
# ==============================================================================
exemplos = [
    "Ol√°, bom dia! Tudo bem?",
    "Acesse https://terrabrasilnoticias.com.br/politica/hoje para saber mais",
    "LIMPAR NOME SPC SERASA AUMENTAR SCORE 100% GARANTIDO", # Deve ser removido
    "Trava zap crash binario 0101010101", # Deve ser removido
    "Veja o v√≠deo no www.youtube.com/watch?v=123"
]

print(">>> Teste do Pre-processamento:")
for msg in exemplos:
    resultado = preprocess_text_advanced(msg)
    status = "REMOVIDO (SPAM/TRAVA)" if not resultado else resultado
    print(f"Original: {msg[:50]}...")
    print(f"Processado: {status}\n")
```

### verificar similaridade

```python
print("\n[ETAPA 2] Gerando representa√ß√£o TF-IDF (com IDF embutido)...")

# garantir inputs
texts = df["clean_text"].astype(str).tolist()
times = df["date_message"].tolist()
N = len(df)

# Vetoriza√ß√£o TF-IDF (IDF embutido)
vectorizer = TfidfVectorizer(min_df=2)
X = vectorizer.fit_transform(texts)

print("Matriz TF-IDF gerada:", X.shape)

# salvar valores de IDF para documenta√ß√£o
idf_df = pd.DataFrame({
    "term": vectorizer.get_feature_names_out(),
    "idf": vectorizer.idf_
}).sort_values(by="idf", ascending=False)

idf_df.to_csv("valores_idf_tfidf.csv", index=False, encoding="utf-8")
print("Tabela de IDF salva em 'valores_idf_tfidf.csv'.")


# ============================================
# 3. FUN√á√ÉO PARA RODAR UM EXPERIMENTO
# ============================================

def run_experiment(k, window_seconds):
    results = []

    for i in range(N - k):      # otimiza√ß√£o
        vec_i = X[i]
        t_i  = times[i]

        for d in range(1, k + 1):
            j = i + d

            t_j = times[j]
            delta_t = (t_j - t_i).total_seconds()

            if delta_t < 0:
                continue
            if delta_t > window_seconds:
                break

            sim_ij = float(vec_i.dot(X[j].T)[0, 0])

            results.append({
                "idx_i": i,
                "idx_j": j,
                "date_i": t_i,
                "date_j": t_j,
                "delta_t_seconds": delta_t,
                "similarity": sim_ij,
                "k": k,
                "window_seconds": window_seconds,
            })

    return pd.DataFrame(results)


# ============================================
# 4. EXPERIMENTOS + CSVs + CCDF + PDF
# ============================================

ks = [2, 5, 10]
windows = [10, 30, 60, 90]
limiar_sim = 0.7
max_dt_pdf = 10

summary_rows = []

def gerar_ccdf(df_exp, k, win):
    df_exp = df_exp.copy()
    df_exp["offset"] = df_exp["idx_j"] - df_exp["idx_i"]
    df_high = df_exp[df_exp["similarity"] >= limiar_sim]

    if df_high.empty:
        print("Nenhum par com similaridade alta para CCDF.")
        return

    plt.figure(figsize=(7,5))

    for offset in range(1, k+1):
        sub = df_high[df_high["offset"] == offset]
        if len(sub) == 0:
            continue

        ts = np.sort(sub["delta_t_seconds"])
        ccdf = 1 - np.arange(1, len(ts)+1) / len(ts)

        plt.loglog(ts, ccdf, label=f"+{offset}")

    plt.xlabel("Time Interval (seconds)")
    plt.ylabel("CCDF")
    plt.title(f"CCDF ‚Äî Similarity ‚â• {limiar_sim} (k={k}, window={win}s)")
    plt.legend()
    plt.tight_layout()
    plt.savefig(f"CCDF_k{k}_win{win}.png", dpi=300)
    plt.close()


def gerar_pdf_similaridade(df_exp, k, win):
    df_short = df_exp[df_exp["delta_t_seconds"] <= max_dt_pdf]
    if df_short.empty:
        print("Nenhum par com Œît <= max_dt_pdf para PDF.")
        return

    sims = df_short["similarity"]
    hist, edges = np.histogram(sims, bins=np.linspace(0,1,20), density=True)
    centers = (edges[:-1] + edges[1:]) / 2

    plt.figure(figsize=(6,4))
    plt.semilogy(centers, hist, marker="o")
    plt.xlabel("Text Similarity")
    plt.ylabel("PDF")
    plt.title(f"PDF ‚Äî Pairs with Œît ‚â§ {max_dt_pdf}s (k={k}, window={win}s)")
    plt.tight_layout()
    plt.savefig(f"PDF_k{k}_win{win}.png", dpi=300)
    plt.close()


# LOOP DOS EXPERIMENTOS
for k in ks:
    for win in windows:
        print(f"\n>>> Rodando experimento k={k}, janela={win}s")
        
        df_exp = run_experiment(k, win)
        print("Pares analisados:", len(df_exp))

        if len(df_exp) == 0:
            continue

        # Estat√≠sticas resumidas
        summary_rows.append({
            "k": k,
            "window_seconds": win,
            "num_pairs": len(df_exp),
            "mean_similarity": df_exp["similarity"].mean(),
            "median_similarity": df_exp["similarity"].median(),
            "p90_similarity": df_exp["similarity"].quantile(0.90),
            "max_similarity": df_exp["similarity"].max(),
        })

        df_exp.to_csv(f"similaridade_k{k}_win{win}s.csv", index=False)
        gerar_ccdf(df_exp, k, win)
        gerar_pdf_similaridade(df_exp, k, win)


# RESUMO FINAL
df_summary = pd.DataFrame(summary_rows)
df_summary.to_csv("resumo_experimentos_similaridade.csv", index=False)
print("\nResumo completo salvo em resumo_experimentos_similaridade.csv")
print(df_summary)
```

```python
import pandas as pd
import scipy.sparse
from tqdm.auto import tqdm  # Barra de progresso

# ==============================================================================
# 1. GARANTIA DE DADOS (Evitar erros de √≠ndice)
# ==============================================================================
# Recarregar listas baseadas no DataFrame atual para garantir sincronia
texts = df["clean_text"].astype(str).tolist()
times = df["date_message"].tolist()
N = len(df)

# Garantir que X (matriz TF-IDF) tenha o mesmo tamanho
if X.shape[0] != N:
    print(f"Recalculando X para garantir tamanho {N}...")
    vectorizer = TfidfVectorizer(min_df=2)
    X = vectorizer.fit_transform(texts)

print(f"Dados prontos: {N} mensagens.")

# ==============================================================================
# 2. FUN√á√ÉO OTIMIZADA (VETORIZADA)
# ==============================================================================
def run_experiment_fast(k, window_seconds):
    results = []
    
    # Barra de progresso para acompanhar
    print(f"Processando k={k}, janela={window_seconds}s...")
    
    # Converter X para CSR se n√£o for (melhora fatiamento por linha)
    X_csr = X.tocsr()
    
    for i in tqdm(range(N - 1), desc="Analisando"):
        t_i = times[i]
        
        # 1. Definir o bloco de vizinhos (slicing √© r√°pido)
        # Pega do pr√≥ximo (i+1) at√© (i+k) - respeitando o limite N
        end_idx = min(i + 1 + k, N)
        if end_idx <= i + 1:
            continue
            
        # Pega a matriz dos vizinhos de uma vez
        neighbors_block = X_csr[i+1 : end_idx]
        
        # 2. C√°lculo Vetorizado de Similaridade
        # Multiplica o vetor i (1xFeatures) pelo bloco de vizinhos (kxFeatures) transposto
        # Resultado √© um vetor (1xk) com todas as similaridades
        # A opera√ß√£o .dot() aqui acontece em C, super r√°pida
        vec_i = X_csr[i]
        sims = neighbors_block.dot(vec_i.T).toarray().flatten()
        
        # 3. Iterar sobre os resultados pr√©-calculados para checar o tempo
        for offset, similarity in enumerate(sims):
            j = i + 1 + offset
            t_j = times[j]
            delta_t = (t_j - t_i).total_seconds()
            
            # Checagem de tempo
            if delta_t < 0:
                continue # ordem temporal invertida (raro se df ordenado)
            if delta_t > window_seconds:
                # Como os vizinhos s√£o sequenciais (d=1, d=2...), se d=1 j√° estourou o tempo,
                # e os dados est√£o ordenados por tempo, podemos parar.
                # Se n√£o estiverem perfeitamente ordenados, melhor n√£o dar break.
                # Assumindo df ordenado por data:
                break 
            
            # Adicionar apenas se v√°lido
            results.append({
                "idx_i": i,
                "idx_j": j,
                "date_i": t_i,
                "date_j": t_j,
                "delta_t_seconds": delta_t,
                "similarity": float(similarity),
                "k": k,
                "window_seconds": window_seconds,
            })

    return pd.DataFrame(results)

# ==============================================================================
# 3. EXECU√á√ÉO
# ==============================================================================
# Rodando com a vers√£o r√°pida
df_full = run_experiment_fast(k=10, window_seconds=90)

print(f"\nTotal de pares encontrados: {len(df_full)}")

if not df_full.empty:
    # Chama a fun√ß√£o de plotagem que definimos anteriormente
    plot_comparative_distributions(df_full)
else:
    print("Nenhum par encontrado nesta janela.")
```

```python
# ================================
# 1. Carregar datasets base
# ================================
df_msgs  = pd.read_csv("dataset_zap_1_preprocessado.csv")
df_coord = pd.read_csv("conjunto_coordenado_mensagens.csv")

# garantir √≠ndice 0..N-1 alinhado com idx_i/idx_j
df_msgs = df_msgs.reset_index(drop=True)
df_msgs["msg_idx"] = df_msgs.index

print("Mensagens totais:", len(df_msgs))
print(df_msgs[["msg_idx", "id_group_anonymous", "id_member_anonymous"]].head())

# ================================
# 2. Grafo de mensagens coordenadas (H)
# ================================
H = nx.Graph()

for _, row in df_coord.iterrows():
    i = int(row["idx_i"])
    j = int(row["idx_j"])
    H.add_edge(i, j)

print("H ‚Äî n√≥s (mensagens):", H.number_of_nodes())
print("H ‚Äî arestas (pares coord.):", H.number_of_edges())

# ================================
# 3. Componentes = cascatas
# ================================
components = list(nx.connected_components(H))
print("N¬∫ de componentes (cascatas):", len(components))

cascade_map = {}
for cid, comp in enumerate(components):
    for m in comp:
        cascade_map[m] = cid

# ================================
# 4. Mensagens coordenadas com cascade_id
# ================================
df_msgs_coord = df_msgs[df_msgs["msg_idx"].isin(cascade_map.keys())].copy()
df_msgs_coord["cascade_id"] = df_msgs_coord["msg_idx"].map(cascade_map)

print("Mensagens coordenadas presentes no dataset:", len(df_msgs_coord))
print(df_msgs_coord[["msg_idx", "cascade_id", "id_group_anonymous", "id_member_anonymous"]].head())

# ================================
# 5. Bipartida Grupo‚ÄìCascata (B_GC)
# ================================
B_GC = nx.Graph()

# n√≥s de grupo
for g in df_msgs_coord["id_group_anonymous"].dropna().unique():
    B_GC.add_node(f"G_{g}", bipartite="group")

# n√≥s de cascata
for c in df_msgs_coord["cascade_id"].dropna().unique():
    B_GC.add_node(f"C_{c}", bipartite="cascade")

# arestas grupo‚Äìcascata
for _, row in df_msgs_coord.iterrows():
    g = row["id_group_anonymous"]
    c = row["cascade_id"]
    if pd.isna(g) or pd.isna(c):
        continue
    B_GC.add_edge(f"G_{g}", f"C_{c}")

print("B_GC ‚Äî n√≥s:", B_GC.number_of_nodes())
print("B_GC ‚Äî arestas:", B_GC.number_of_edges())

# ================================
# 6. Bipartida Usu√°rio‚ÄìCascata (B_UC)
# ================================
B_UC = nx.Graph()

# n√≥s de usu√°rio
for u in df_msgs_coord["id_member_anonymous"].dropna().unique():
    B_UC.add_node(f"U_{u}", bipartite="user")

# n√≥s de cascata
for c in df_msgs_coord["cascade_id"].dropna().unique():
    B_UC.add_node(f"C_{c}", bipartite="cascade")

# arestas user‚Äìcascade
for _, row in df_msgs_coord.iterrows():
    u = row["id_member_anonymous"]
    c = row["cascade_id"]
    if pd.isna(u) or pd.isna(c):
        continue
    B_UC.add_edge(f"U_{u}", f"C_{c}")

print("B_UC ‚Äî n√≥s:", B_UC.number_of_nodes())
print("B_UC ‚Äî arestas:", B_UC.number_of_edges())

# ================================
# 7. Proje√ß√µes: GCN_coord (grupo) e UCN_coord (usu√°rio)
# ================================
from networkx.algorithms import bipartite

# GCN_coord
group_nodes = [n for n, d in B_GC.nodes(data=True) if d.get("bipartite") == "group"]
GCN_coord = bipartite.weighted_projected_graph(B_GC, group_nodes)

print("GCN_coord ‚Äî grupos:", GCN_coord.number_of_nodes())
print("GCN_coord ‚Äî arestas:", GCN_coord.number_of_edges())

# UCN_coord
user_nodes = [n for n, d in B_UC.nodes(data=True) if d.get("bipartite") == "user"]
UCN_coord = bipartite.weighted_projected_graph(B_UC, user_nodes)

print("UCN_coord ‚Äî usu√°rios:", UCN_coord.number_of_nodes())
print("UCN_coord ‚Äî arestas:", UCN_coord.number_of_edges())
```

```python
def get_degree_series(G, weighted=False):
    """
    Retorna uma Series com os graus (simples ou ponderado).
    """
    if weighted:
        deg = dict(G.degree(weight="weight"))
    else:
        deg = dict(G.degree())
    s = pd.Series(deg, name="degree")
    return s

def get_weight_series(G):
    """
    Retorna uma Series com os pesos das arestas.
    """
    weights = [d.get("weight", 1.0) for _, _, d in G.edges(data=True)]
    return pd.Series(weights, name="weight")

def plot_pdf_ccdf(series, title_prefix, file_prefix, loglog_ccdf=True):
    """
    Plota PDF e CCDF da s√©rie (graus ou pesos).
    """
    s = series[series > 0].astype(float)

    # PDF
    plt.figure()
    plt.hist(s, bins=30, density=True)
    plt.xlabel("Value")
    plt.ylabel("PDF")
    plt.title(f"{title_prefix} ‚Äî PDF")
    plt.tight_layout()
    plt.savefig(f"{file_prefix}_PDF.png", dpi=300)
    plt.close()

    # CCDF
    vals = np.sort(s.values)
    n = len(vals)
    ccdf = 1.0 - np.arange(1, n + 1) / n

    plt.figure()
    if loglog_ccdf:
        plt.loglog(vals, ccdf, marker=".")
    else:
        plt.plot(vals, ccdf, marker=".")
    plt.xlabel("Value")
    plt.ylabel("CCDF")
    plt.title(f"{title_prefix} ‚Äî CCDF")
    plt.tight_layout()
    plt.savefig(f"{file_prefix}_CCDF.png", dpi=300)
    plt.close()

    print(f"Salvo: {file_prefix}_PDF.png e {file_prefix}_CCDF.png")
```

### An√°lise estrutural das cascatas coordenadas

```python
import pandas as pd
import networkx as nx
from tqdm import tqdm
import matplotlib.pyplot as plt
import numpy as np

# ===========================
# 1. Carregar dados
# ===========================
df_coord = pd.read_csv("conjunto_coordenado_mensagens.csv")
df_msgs  = pd.read_csv("dataset_zap_1_preprocessado.csv")

# MUITO IMPORTANTE: o √≠ndice do df_msgs deve coincidir com os idx_i/idx_j
# usados nos experimentos de similaridade.
# Ou seja: √≠ndice 0..N-1 na mesma ordem do df usado no TF-IDF.
df_msgs = df_msgs.reset_index(drop=True)

# s√≥ pra conferir:
print("Tamanho df_msgs:", len(df_msgs))
print("idx_i min/max:", df_coord["idx_i"].min(), df_coord["idx_i"].max())
print("idx_j min/max:", df_coord["idx_j"].min(), df_coord["idx_j"].max())

# Se o max(idx_j) < len(df_msgs), est√° alinhado.

# Vamos usar o √çNDICE como chave (n√£o coluna id_message)
# to_dict() usa o √≠ndice como chave automaticamente
msg_time  = df_msgs["date_message"].to_dict()
msg_user  = df_msgs["id_member_anonymous"].to_dict()
msg_group = df_msgs["id_group_anonymous"].to_dict()

# ===========================
# 2. Construir grafo de mensagens coordenadas
# ===========================
G = nx.Graph()
for _, row in df_coord.iterrows():
    i = int(row["idx_i"])
    j = int(row["idx_j"])
    G.add_edge(i, j, similarity=row["similarity"])

print("N√≥s na rede:", G.number_of_nodes())
print("Arestas na rede:", G.number_of_edges())

# ===========================
# 3. Componentes (cascatas)
# ===========================
components = sorted(nx.connected_components(G), key=len, reverse=True)
print("N√∫mero de cascatas:", len(components))

casc_stats = []

for cid, comp in enumerate(tqdm(components)):
    comp = list(comp)
    subG = G.subgraph(comp)

    # tamanho estrutural
    size = len(subG)
    edges = subG.number_of_edges()

    # graus (hubs)
    degs = dict(subG.degree())
    hub = max(degs, key=degs.get)
    hub_degree = degs[hub]

    # --- temporal (usando √≠ndice do DataFrame) ---
    # se por alguma raz√£o houver algum m que n√£o esteja em msg_time,
    # a gente descarta (s√≥ por seguran√ßa extra)
    valid_msgs = [m for m in comp if m in msg_time]

    if not valid_msgs:
        # se cair aqui, algo est√° MUITO desalinhado ‚Äî voc√™ pode logar para inspecionar
        continue

    times = [pd.to_datetime(msg_time[m]) for m in valid_msgs]
    t_min = min(times)
    t_max = max(times)
    duration = (t_max - t_min).total_seconds()

    # grupos e usu√°rios envolvidos
    groups = list({msg_group[m] for m in valid_msgs if m in msg_group})
    users  = list({msg_user[m]  for m in valid_msgs if m in msg_user})

    casc_stats.append({
        "cascade_id": cid,
        "num_messages": size,
        "num_edges": edges,
        "hub_message": hub,
        "hub_user": msg_user.get(hub, None),
        "hub_degree": hub_degree,
        "num_users": len(users),
        "num_groups": len(groups),
        "start_time": t_min,
        "end_time": t_max,
        "duration_seconds": duration
    })

df_casc = pd.DataFrame(casc_stats)
df_casc.to_csv("cascatas_estatisticas.csv", index=False)
df_casc.head()
```

```python
def plot_community_network(
    nodes_csv,
    edges_csv,
    id_col,        # "group_node" ou "user_id"
    src_col,       # "group_u"   ou "user_u"
    dst_col,       # "group_v"   ou "user_v"
    title,
    out_png,
    top_n=None     # se quiser plotar s√≥ os top hubs (ex: 500)
):
    # 1. Carregar dados
    df_nodes = pd.read_csv(nodes_csv)
    df_edges = pd.read_csv(edges_csv)
    
    # 2. (Opcional) filtrar s√≥ top_n n√≥s por grau ponderado
    if top_n is not None and len(df_nodes) > top_n:
        keep_ids = set(
            df_nodes.sort_values("degree_weighted", ascending=False)
                    .head(top_n)[id_col]
        )
        df_nodes = df_nodes[df_nodes[id_col].isin(keep_ids)]
        df_edges = df_edges[
            df_edges[src_col].isin(keep_ids) &
            df_edges[dst_col].isin(keep_ids)
        ]
    
    # 3. Construir grafo
    G = nx.Graph()
    
    # adicionar n√≥s com todos os atributos do CSV
    for _, row in df_nodes.iterrows():
        node_id = row[id_col]
        attrs = row.to_dict()
        G.add_node(node_id, **attrs)
    
    # adicionar arestas
    for _, row in df_edges.iterrows():
        u = row[src_col]
        v = row[dst_col]
        w = row.get("weight", 1.0)
        if u in G.nodes and v in G.nodes:
            G.add_edge(u, v, weight=w)
    
    print(f"Grafo para {title}")
    print("  N√≥s:", G.number_of_nodes())
    print("  Arestas:", G.number_of_edges())
    
    # 4. Layout
    # k controla a ‚Äúespalhada‚Äù do grafo. Ajuste se ficar muito junto.
    pos = nx.spring_layout(G, k=0.15, iterations=100, seed=42)
    
    # 5. Cores por comunidade
    communities = []
    for n, data in G.nodes(data=True):
        # se n√£o tiver community, coloca -1
        communities.append(int(data.get("community", -1)))
    
    unique_coms = sorted(set(communities))
    cmap = cm.get_cmap("tab20", len(unique_coms))
    color_map = {c: cmap(i) for i, c in enumerate(unique_coms)}
    node_colors = [color_map[c] for c in communities]
    
    # 6. Tamanho dos n√≥s pelo degree_weighted (transforma√ß√£o suave)
    sizes = []
    for n, data in G.nodes(data=True):
        dw = data.get("degree_weighted", 1)
        # raiz ou log pra n√£o explodir
        sizes.append(40 + 5 * np.sqrt(dw))
    
    # 7. Largura das arestas pelo peso
    edge_widths = []
    for u, v, data in G.edges(data=True):
        w = data.get("weight", 1)
        edge_widths.append(0.3 + np.log1p(w))  # log(1+w)
    
    # 8. Plot
    plt.figure(figsize=(10, 10))
    
    # arestas primeiro
    nx.draw_networkx_edges(
        G, pos,
        alpha=0.08,
        width=edge_widths,
        edge_color="gray"
    )
    
    # n√≥s
    nx.draw_networkx_nodes(
        G, pos,
        node_size=sizes,
        node_color=node_colors,
        linewidths=0.1,
        edgecolors="black",
        alpha=0.9
    )
    
    plt.title(title)
    plt.axis("off")
    plt.tight_layout()
    plt.savefig(out_png, dpi=300, bbox_inches="tight")
    plt.show()
    print(f"Figura salva em: {out_png}")
```

```python
def visualizar_rede_memoria(G, df_metrics, node_col_id, title, output_filename, top_k=None):
    """
    Plota a rede usando o objeto NetworkX (G) e o DataFrame de m√©tricas (para cores/tamanhos).
    """
    plt.figure(figsize=(12, 10))
    
    # 1. Filtragem (se a rede for muito grande, plotamos apenas os 'top_k' hubs)
    if top_k and G.number_of_nodes() > top_k:
        # Pegar os top_k n√≥s com maior grau ponderado
        top_nodes = set(df_metrics.sort_values("degree_weighted", ascending=False).head(top_k)[node_col_id])
        H = G.subgraph(top_nodes)
        print(f"üîç Filtrando rede: exibindo os {top_k} principais n√≥s (de {G.number_of_nodes()} originais).")
    else:
        H = G
        print(f"üîç Exibindo rede completa: {H.number_of_nodes()} n√≥s.")

    # 2. Layout (Spring Layout para for√ßar clusteriza√ß√£o visual)
    print("‚è≥ Calculando layout (pode demorar um pouco)...")
    pos = nx.spring_layout(H, k=0.15, seed=42, iterations=50)

    # 3. Preparar Cores (Comunidades) e Tamanhos (Grau)
    # Mapear atributos do dataframe para o subgrafo H
    communities = []
    sizes = []
    
    # Dicion√°rio auxiliar para busca r√°pida
    meta_dict = df_metrics.set_index(node_col_id).to_dict('index')

    for node in H.nodes():
        if node in meta_dict:
            comm = meta_dict[node].get('community', 0)
            deg = meta_dict[node].get('degree_weighted', 1)
        else:
            comm = 0
            deg = 1
        
        communities.append(comm)
        sizes.append(deg)

    # Normalizar tamanhos para n√£o ficarem gigantes
    sizes = np.array(sizes)
    # Logaritmo suaviza a diferen√ßa entre hubs gigantes e n√≥s pequenos
    node_sizes = 50 + (np.log1p(sizes) * 30) 

    # 4. Desenhar
    # Arestas
    nx.draw_networkx_edges(H, pos, alpha=0.1, edge_color='#999999')
    
    # N√≥s
    # Usar colormap 'tab20' para comunidades distintas
    nx.draw_networkx_nodes(
        H, pos, 
        node_size=node_sizes, 
        node_color=communities, 
        cmap=plt.cm.tab20, 
        alpha=0.9,
        linewidths=0.5,
        edgecolors='white'
    )

    # Labels (apenas para os top 5 do plot para n√£o poluir)
    if top_k:
        # Identificar os 5 maiores no subgrafo atual para rotular
        top_5_local = sorted(list(H.nodes()), key=lambda n: meta_dict.get(n, {}).get('degree_weighted', 0), reverse=True)[:5]
        labels = {n: n for n in top_5_local}
        nx.draw_networkx_labels(H, pos, labels, font_size=8, font_color='black', font_weight='bold')

    plt.title(f"{title}\n(Cor = Comunidade, Tamanho = Grau Ponderado)", fontsize=14)
    plt.axis('off')
    
    # Salvar
    plt.savefig(output_filename, dpi=300, bbox_inches='tight')
    plt.show()
    print(f"‚úÖ Visualiza√ß√£o salva em: {output_filename}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
from nltk.util import ngrams

# 1. Preparar os dados
# Carregar m√©tricas de usu√°rios se n√£o estiverem em mem√≥ria
if 'df_nodes_u' not in locals():
    df_nodes_u = pd.read_csv("UCN_coord_nodes_metricas_comunidades.csv")

# Precisamos garantir que df_msgs tem as mensagens.
# Se df_msgs n√£o estiver na mem√≥ria, recarregue o preprocessado:
if 'df_msgs' not in locals():
    df_msgs = pd.read_csv("dataset_zap_1_preprocessado.csv")

# Merge: Adicionar a coluna 'community' ao dataframe de mensagens
# Ligamos pelo ID do usu√°rio
df_analise_semantica = pd.merge(
    df_msgs[['id_member_anonymous', 'clean_text', 'date_message']], 
    df_nodes_u[['user_id', 'community', 'degree_weighted']], 
    left_on='id_member_anonymous', 
    right_on='user_id',
    how='inner'
)

print(f"Mensagens vinculadas a usu√°rios coordenados: {len(df_analise_semantica)}")

# 2. Identificar as maiores comunidades
top_communities = df_analise_semantica['community'].value_counts().head(6).index.tolist()
print(f"Analisando as Top 6 Comunidades: {top_communities}")

# 3. Fun√ß√£o para plotar os termos
def plotar_top_termos(df, community_id, ax_uni, ax_bi):
    # Filtrar mensagens da comunidade
    msgs = df[df['community'] == community_id]['clean_text'].dropna().astype(str).tolist()
    
    # Tokeniza√ß√£o simples (j√° est√° limpo, ent√£o split por espa√ßo funciona bem)
    all_tokens = []
    for text in msgs:
        tokens = text.split()
        # Filtro extra de tamanho para garantir
        tokens = [t for t in tokens if len(t) > 2] 
        all_tokens.extend(tokens)
        
    # Unigramas (Palavras soltas)
    counter_uni = Counter(all_tokens)
    most_common_uni = counter_uni.most_common(10)
    
    # Bigramas (Pares de palavras - d√£o mais contexto)
    bi_grams = list(ngrams(all_tokens, 2))
    counter_bi = Counter([' '.join(g) for g in bi_grams])
    most_common_bi = counter_bi.most_common(10)
    
    # Plot Unigramas
    if most_common_uni:
        words, counts = zip(*most_common_uni)
        y_pos = range(len(words))
        ax_uni.barh(y_pos, counts, align='center', color='skyblue')
        ax_uni.set_yticks(y_pos)
        ax_uni.set_yticklabels(words)
        ax_uni.invert_yaxis()
        ax_uni.set_title(f"Comunidade {community_id}: Top Palavras")
    
    # Plot Bigramas
    if most_common_bi:
        words, counts = zip(*most_common_bi)
        y_pos = range(len(words))
        ax_bi.barh(y_pos, counts, align='center', color='salmon')
        ax_bi.set_yticks(y_pos)
        ax_bi.set_yticklabels(words)
        ax_bi.invert_yaxis()
        ax_bi.set_title(f"Comunidade {community_id}: Top Bigramas")

# 4. Loop de Visualiza√ß√£o
fig, axes = plt.subplots(len(top_communities), 2, figsize=(16, 4 * len(top_communities)))
plt.subplots_adjust(hspace=0.4, wspace=0.3)

for i, com_id in enumerate(top_communities):
    if len(top_communities) > 1:
        ax_uni = axes[i, 0]
        ax_bi = axes[i, 1]
    else:
        ax_uni = axes[0]
        ax_bi = axes[1]
        
    plotar_top_termos(df_analise_semantica, com_id, ax_uni, ax_bi)

plt.show()
```

```python
import pandas as pd

# 1. Carregar m√©tricas dos Grupos (GCN)
# Este arquivo foi gerado na c√©lula 31 do notebook original
try:
    df_nodes_g = pd.read_csv("GCN_coord_nodes_metricas_comunidades.csv")
except:
    print("‚ö†Ô∏è Arquivo de m√©tricas de grupo n√£o encontrado. Rode a c√©lula 31 do notebook original.")

# 2. Top Grupos por Import√¢ncia (Grau e Intermedia√ß√£o)
# Degree Weighted: Grupos onde ocorre MUITA coordena√ß√£o (Alvos principais)
# Betweenness: Grupos que ligam diferentes comunidades (Pontes de cont√°gio)

top_groups = df_nodes_g.sort_values("degree_weighted", ascending=False).head(10).copy()

# Adicionar um rank leg√≠vel
top_groups['Rank_Grau'] = range(1, 11)

print("üèÜ Top 10 Grupos (Palcos da Coordena√ß√£o):")
display(top_groups[['group_id', 'community', 'degree_weighted', 'betweenness']])

# Pegar os IDs para an√°lise profunda
top_group_ids = top_groups['group_id'].tolist()
```

```python
from collections import Counter

# 1. Merge Mensagens com Grupos
# Se df_msgs n√£o estiver na mem√≥ria, recarregue: pd.read_csv("dataset_zap_1_preprocessado.csv")
if 'df_msgs' in locals():
    # Filtrar apenas mensagens dos top grupos para ser r√°pido
    msgs_top_groups = df_msgs[df_msgs['id_group_anonymous'].isin(top_group_ids[:5])]

    print(f"Analisando conte√∫do dos 5 maiores grupos...")

    for gid in top_group_ids[:5]:
        # Pegar textos do grupo
        texts = msgs_top_groups[msgs_top_groups['id_group_anonymous'] == gid]['clean_text'].dropna().astype(str)
        
        # Tokeniza√ß√£o simples
        all_words = " ".join(texts).split()
        all_words = [w for w in all_words if len(w) > 3] # Filtra palavras curtas
        
        # Top 5 palavras
        top_words = Counter(all_words).most_common(7)
        words_str = ", ".join([w[0] for w in top_words])
        
        # Pegar a comunidade do grupo
        comm = df_nodes_g[df_nodes_g['group_id'] == gid]['community'].values[0]
        
        print(f"Grupo {gid[-6:]}... (Comunidade {comm})")
        print(f"   Temas: {words_str}")
        print("-" * 40)
```

```python
import matplotlib.pyplot as plt

# 1. Preparar dados
# Usamos a coluna 'estado_limpo' que criamos na etapa de usu√°rios (se existir), ou recriamos
if 'estado_limpo' not in df_msgs.columns:
    # (L√≥gica simplificada se voc√™ pulou a etapa anterior)
    df_msgs['estado_limpo'] = df_msgs['state'].fillna('Desconhecido')

# 2. Filtrar dados dos Top 5 Grupos
df_groups_geo = df_msgs[df_msgs['id_group_anonymous'].isin(top_group_ids[:5])]

# 3. Plotar
plt.figure(figsize=(14, 6))

# Tabela cruzada: Grupo x Estado
ct = pd.crosstab(df_groups_geo['id_group_anonymous'], df_groups_geo['estado_limpo'])

# Normalizar para ver porcentagem (melhor para comparar grupos de tamanhos diferentes)
ct_norm = ct.div(ct.sum(1), axis=0)

# Plotar apenas estados relevantes (que tenham pelo menos 5% em algum grupo)
cols_to_plot = ct_norm.columns[ct_norm.max() > 0.05]
ct_norm[cols_to_plot].plot(kind='bar', stacked=True, figsize=(14, 6), colormap='tab20')

plt.title("Composi√ß√£o Geogr√°fica dos Principais Grupos (Onde est√£o os membros?)")
plt.ylabel("Propor√ß√£o de Usu√°rios (%)")
plt.xlabel("ID do Grupo (Anonimizado)")
# Simplificar labels do eixo X para n√£o ficar gigante
plt.xticks(ticks=range(5), labels=[gid[-6:] for gid in top_group_ids[:5]], rotation=0)
plt.legend(bbox_to_anchor=(1.01, 1), loc='upper left', title="Estado")
plt.tight_layout()
plt.show()
```

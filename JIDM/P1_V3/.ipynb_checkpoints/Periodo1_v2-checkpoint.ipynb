```python
print(">>> [REFINAMENTO FINAL V3] Removendo IPTV (Popcorn), Internet (Antenna) e Harbyngers...")

# 1. LISTAS DE BLOQUEIO V3 (CirÃºrgicas)
# ------------------------------------------------------------------------------

# Vandalismo e Travas
crash_keywords = [
    "trava", "crash", "binario", "destrava", "imune", "rajada", 
    "virus", "clonagem", "hacker", "phoenix", "revaneio", "alien_monster", 
    "à¿—", "ê™°", "â€®", "bÃ²sta nÃ ", "carÃ¡Ãª", 
    "auto entitula logan", "raimundo",
    # NOVOS (O que vazou):
    "harbyngers", "fish", "crown", "sun" # Assinatura do trava que sobrou
]

# Spam Comercial
spam_keywords = [
    "limpar nome", "aumentar score", "cnpj", "crÃ©dito", "emprÃ©stimo", 
    "venda", "oferta", "desconto", "turismo", "excursÃ£o", "saÃ­da", 
    "ganhe", "oportunidade", "renda extra", "marketing", "consÃ³rcio",
    "delivery", "entregamos", "promoÃ§Ã£o",
    # NOVOS (O que vazou - IPTV e Internet):
    "popcorn", "clapper_board", "antenna_bars", # Emojis de spam
    "melhores canais", "sua tv", "venha adquirir", "adquirir sua",
    "internet ilimitada", "conexÃ£o ilimitada", "sem travamentos",
    "filmes", "sÃ©ries", "canais", "iptv", "tv box", "hbo", "telecine", 
    "INTERNET ILIMITADA", "Internet Ilimitada", "vivo", "tim", "oi", "Venha"
]

# Stopwords
custom_stopwords = {
    "bom", "dia", "boa", "tarde", "noite", "olÃ¡", "oi", "ola", "tchau",
    "kk", "kkk", "kkkk", "kkkkk", "rs", "rsrs", "aff", "eita", "oxi", "oxe",
    "imagem", "figura", "sticker", "Ã¡udio", "audio", "vÃ­deo", "video",
    "mensagem", "encaminhada", "apagou"
}
# stop_words_total = custom_stopwords # Descomente se precisar

# ==============================================================================
# 2. FUNÃ‡Ã•ES AUXILIARES
# ==============================================================================

def is_junk_message(text_lower):
    """
    Retorna True se a mensagem for considerada Lixo (Spam, Trava ou Vandalismo)
    """
    # 1. Checagem de Travazap por tamanho excessivo
    # Mensagens legÃ­timas de WhatsApp raramente passam de 3000 caracteres
    if len(text_lower) > 3000:
        return True
        
    # 2. Checagem por palavras-chave de Vandalismo
    if any(term in text_lower for term in crash_keywords):
        return True
        
    # 3. Checagem por SPAM Comercial
    if any(term in text_lower for term in spam_keywords):
        return True
        
    return False

def simplify_urls(text):
    """
    Substitui URLs completas apenas pelo domÃ­nio principal.
    Ex: 'https://terrabrasilnoticias.com/2022/noticia' -> 'terrabrasilnoticias.com'
    """
    # Regex para encontrar URLs (http, https ou www)
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    
    def extract_domain(match):
        url = match.group(0)
        # Adiciona http se faltar para o urlparse funcionar
        if not url.startswith('http'):
            url = 'http://' + url
        try:
            parsed = urlparse(url)
            domain = parsed.netloc
            # Remove o 'www.' se existir para limpar mais
            return domain.replace('www.', '')
        except:
            return url # Se falhar, devolve o original

    return url_pattern.sub(extract_domain, text)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from bisect import bisect_right
from tqdm.auto import tqdm

# ==============================================================================
# 1. SONDAGEM AMPLIADA (1000 SEGUNDOS)
# ==============================================================================
print(">>> [SONDAGEM] Coletando dados (Janela 1000s)...")

# Precisamos garantir que a funÃ§Ã£o pegue atÃ© 1000s
def coletar_dados_sondagem_1000s(max_window=1000, min_sim=0.70):
    amostras = []
    time_window = pd.Timedelta(seconds=max_window)
    
    # Amostragem
    step = 1 if N < 30000 else 2 
    
    for i in tqdm(range(0, N - 1, step), desc="Sondando"):
        t_i = times[i]
        t_limit = t_i + time_window
        
        end_idx = bisect_right(times, t_limit, lo=i+1)
        if end_idx <= i + 1: continue
            
        block = X[i+1 : end_idx]
        vec_i = X[i]
        sims = block.dot(vec_i.T).toarray().flatten()
        
        # Filtro
        valid_indices = np.where(sims >= min_sim)[0]
        
        if len(valid_indices) > 0:
            global_indices = (i + 1) + valid_indices
            valid_sims = sims[valid_indices]
            for g_idx, s_val in zip(global_indices, valid_sims):
                dt = (times[g_idx] - t_i).total_seconds()
                amostras.append({"delta_t": dt, "similarity": float(s_val)})

    return pd.DataFrame(amostras)

# Rodar a coleta
df_sondagem = coletar_dados_sondagem_1000s()

# ==============================================================================
# 2. PLOTAGEM CORRIGIDA (BINS vs LABELS)
# ==============================================================================
if not df_sondagem.empty:
    plt.rcParams.update({'font.size': 12, 'axes.linewidth': 1.2, 'grid.alpha': 0.3})

    # --- GRÃFICO 1: PDF (Densidade) ---
    plt.figure(figsize=(10, 6))
    
    # CORREÃ‡ÃƒO AQUI:
    # 5 pontos de corte = 4 faixas
    bins = [0, 10, 60, 300, 1000]
    labels = [
        '0-10s (Imediato)', 
        '10-60s (Curto)', 
        '60-300s (MÃ©dio)', 
        '300-1000s (Longo)'
    ]
    
    # Agora vai funcionar porque len(labels) == len(bins) - 1
    df_sondagem['faixa_tempo'] = pd.cut(df_sondagem['delta_t'], bins=bins, labels=labels)
    
    sns.kdeplot(
        data=df_sondagem, 
        x='similarity', 
        hue='faixa_tempo', 
        fill=True, 
        common_norm=False, 
        palette="viridis",
        alpha=0.1
    )
    
    plt.xlim(0.7, 1.0) # Foco na alta similaridade
    plt.title('PDF: DistribuiÃ§Ã£o de Similaridade por Tempo (AtÃ© 1000s)')
    plt.xlabel('Similaridade')
    plt.ylabel('Densidade')
    plt.grid(True, linestyle='--')
    plt.savefig("Grafico_Sondagem_1000s_PDF.png", dpi=300)
    plt.show()

    # --- GRÃFICO 2: CDF (AcumulaÃ§Ã£o) ---
    plt.figure(figsize=(9, 6))
    
    limiar_sim = 0.90
    df_high = df_sondagem[df_sondagem['similarity'] >= limiar_sim]
    
    if not df_high.empty:
        sorted_data = np.sort(df_high['delta_t'])
        yvals = np.arange(1, len(sorted_data) + 1) / len(sorted_data)
        
        plt.step(sorted_data, yvals, where='post', color='#ff7f0e', linewidth=3, label=rf'Sim $\geq$ {limiar_sim}')
        
        # MarcaÃ§Ãµes importantes
        plt.axvline(60, color='#1f77b4', linestyle='--', label='60s')
        plt.axvline(300, color='gray', linestyle=':', label='300s')
        
        # Calcular %
        perc_60 = len(df_high[df_high['delta_t'] <= 60]) / len(df_high)
        plt.text(65, perc_60 - 0.05, f"{perc_60:.1%} em 60s", color='#1f77b4', fontweight='bold')

        plt.xlim(0, 1000) # Mostra atÃ© 1000s agora
        plt.ylim(0, 1.05)
        
    plt.xlabel("Tempo (segundos)")
    plt.ylabel("ProporÃ§Ã£o Acumulada")
    plt.title(f"CDF (Janela Extendida 1000s) - Sim >= {limiar_sim}")
    plt.legend()
    plt.grid(True, linestyle='--')
    plt.savefig("Grafico_Sondagem_1000s_CDF.png", dpi=300)
    plt.show()

else:
    print("Nenhum dado encontrado na sondagem.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from bisect import bisect_right
from tqdm.auto import tqdm

# 1. GARANTIA: Atualizar X e times com o dataset limpo atual
# (Caso vocÃª tenha rodado cÃ©lulas fora de ordem)
print(">>> [RE-VETORIZAÃ‡ÃƒO] Atualizando matriz com texto limpo...")
from sklearn.feature_extraction.text import TfidfVectorizer
import scipy.sparse

# Atualiza listas
texts = df["clean_text"].astype(str).tolist()
times = df["date_message"].tolist()
N = len(df)

# Gera TF-IDF novo
vectorizer = TfidfVectorizer(min_df=2) 
X = vectorizer.fit_transform(texts)
X = X.tocsr()
print(f"Matriz atualizada: {X.shape}")

# ==============================================================================
# 2. SONDAGEM (Coleta de dados para os grÃ¡ficos)
# ==============================================================================
print("\n>>> [SONDAGEM] Coletando dados para os grÃ¡ficos (Janela 300s)...")

def coletar_dados_sondagem(max_window=300, min_sim=0.70):
    amostras = []
    time_window = pd.Timedelta(seconds=max_window)
    
    # Amostragem inteligente: se for muito grande, pula alguns Ã­ndices para ser rÃ¡pido
    step = 1 if N < 30000 else 2 
    
    for i in tqdm(range(0, N - 1, step), desc="Sondando"):
        t_i = times[i]
        t_limit = t_i + time_window
        
        end_idx = bisect_right(times, t_limit, lo=i+1)
        if end_idx <= i + 1: continue
            
        block = X[i+1 : end_idx]
        vec_i = X[i]
        sims = block.dot(vec_i.T).toarray().flatten()
        
        # Pega tudo acima de 0.70
        valid_indices = np.where(sims >= min_sim)[0]
        
        if len(valid_indices) > 0:
            global_indices = (i + 1) + valid_indices
            valid_sims = sims[valid_indices]
            for g_idx, s_val in zip(global_indices, valid_sims):
                dt = (times[g_idx] - t_i).total_seconds()
                amostras.append({"delta_t": dt, "similarity": float(s_val)})

    return pd.DataFrame(amostras)

df_sondagem = coletar_dados_sondagem()

# ==============================================================================
# 3. GERAR GRÃFICOS (PDF + CDF)
# ==============================================================================
if not df_sondagem.empty:
    # ConfiguraÃ§Ã£o Visual
    plt.rcParams.update({'font.size': 12, 'axes.linewidth': 1.2, 'grid.alpha': 0.3})

    # --- GRÃFICO 1: PDF (Densidade de Similaridade) ---
    plt.figure(figsize=(9, 6))
    subset = df_sondagem[df_sondagem['delta_t'] <= 60]
    
    if not subset.empty:
        bins = np.linspace(0, 1, 21)
        counts, edges = np.histogram(subset['similarity'], bins=bins, density=True)
        centers = (edges[:-1] + edges[1:]) / 2
        
        plt.semilogy(centers, counts, marker='o', linestyle='-', linewidth=2.5, 
                     markersize=9, color='#1f77b4', label=r'Pairs with $\Delta t \leq 60s$')

    plt.xlabel("Text Similarity")
    plt.ylabel("PDF (Log Scale)")
    plt.title("PDF â€” Similarity Distribution")
    plt.grid(True, which="major", linestyle='--')
    plt.legend()
    plt.tight_layout()
    plt.savefig("Grafico_Definitivo_PDF.png", dpi=300)
    plt.show()

    # --- GRÃFICO 2: CDF (AcumulaÃ§Ã£o no Tempo) ---
    plt.figure(figsize=(9, 6))
    limiar_sim = 0.90 # Vamos testar o rigor de 0.90 agora
    df_high = df_sondagem[df_sondagem['similarity'] >= limiar_sim]
    
    if not df_high.empty:
        sorted_data = np.sort(df_high['delta_t'])
        yvals = np.arange(1, len(sorted_data) + 1) / len(sorted_data)
        
        plt.step(sorted_data, yvals, where='post', color='#ff7f0e', linewidth=3, label=rf'Sim $\geq$ {limiar_sim}')
        plt.axvline(60, color='#1f77b4', linestyle='--', linewidth=1.5, label='Cutoff 60s')
        
        # AnotaÃ§Ã£o de %
        perc = len(df_high[df_high['delta_t'] <= 60]) / len(df_high)
        plt.plot(60, perc, 'o', color='#1f77b4', markersize=8)
        plt.text(65, perc - 0.1, f"{perc:.1%} accumulated\nin first 60s", color='#1f77b4', fontweight='bold')

        plt.xlim(0, 180)
        plt.ylim(0, 1.05)
    
    plt.xlabel("Time Interval (seconds)")
    plt.ylabel("CDF (Cumulative Distribution)")
    plt.title(rf"CDF â€” Coordination Speed (Sim $\geq$ {limiar_sim})")
    plt.grid(True, linestyle='--')
    plt.legend()
    plt.tight_layout()
    plt.savefig("Grafico_Definitivo_CDF.png", dpi=300)
    plt.show()
    
    print("\n>>> ANÃLISE PRONTA!")
    print("Verifique os grÃ¡ficos acima. Se o CDF mostra que a maioria acontece em <60s,")
    print("vocÃª pode rodar o BLOCO 2 (Motor de Similaridade) com janela=60s e sim=0.90.")

else:
    print("Nenhum dado encontrado. Verifique se o dataset nÃ£o ficou vazio.")
```

```python
# ==============================================================================
# BLOCO 2: MOTOR DE SIMILARIDADE (Janela Temporal + Bisect)
# ==============================================================================
import numpy as np
import pandas as pd
from bisect import bisect_right
from tqdm.auto import tqdm

def processar_janela_temporal(window_seconds, min_sim):
    print(f"\n>>> Iniciando busca: Janela={window_seconds}s | Sim={min_sim}")
    results = []
    
    # Prepara a janela de tempo para soma direta
    time_window = pd.Timedelta(seconds=window_seconds)
    
    # Barra de progresso varrendo todas as mensagens (0 atÃ© N-1)
    for i in tqdm(range(N - 1), desc="Processando Janelas"):
        t_i = times[i]
        t_limit = t_i + time_window
        
        # 1. Busca BinÃ¡ria: Acha onde termina a janela de tempo instantaneamente
        # Procura na lista 'times' o Ã­ndice da Ãºltima mensagem que cabe na janela
        end_idx = bisect_right(times, t_limit, lo=i+1)
        
        # Se nÃ£o tiver vizinhos na janela, avanÃ§a
        if end_idx <= i + 1:
            continue
            
        # 2. ExtraÃ§Ã£o do Bloco (Slicing da Matriz X)
        # Pega a matriz dos vizinhos (do prÃ³ximo atÃ© o limite do tempo)
        block_neighbors = X[i+1 : end_idx]
        vec_i = X[i]
        
        # 3. CÃ¡lculo Vetorizado (Produto Escalar)
        # Calcula similaridade de i contra todos os vizinhos de uma vez
        sims = block_neighbors.dot(vec_i.T).toarray().flatten()
        
        # 4. Filtragem por Similaridade
        # Pega apenas os Ã­ndices que superaram o limiar (ex: 0.90)
        valid_indices = np.where(sims >= min_sim)[0]
        
        if len(valid_indices) > 0:
            # Recupera Ã­ndices globais e valores
            global_indices = (i + 1) + valid_indices
            valid_sims = sims[valid_indices]
            
            # Adiciona Ã  lista de resultados
            # Usamos list comprehension para ser mais rÃ¡pido
            for global_idx, sim_val in zip(global_indices, valid_sims):
                results.append({
                    "source": i,
                    "target": global_idx,
                    "time_i": t_i,
                    "delta_sec": (times[global_idx] - t_i).total_seconds(),
                    "weight": float(sim_val)
                })

    return pd.DataFrame(results)

# --- EXECUÃ‡ÃƒO ---
# Aqui definimos os parÃ¢metros da tese
df_edges_bruto = processar_janela_temporal(window_seconds=60, min_sim=0.70)

if not df_edges_bruto.empty:
    df_edges_bruto.to_csv("arestas_brutas_temporal.csv", index=False, sep=";")
    print(f"Sucesso! {len(df_edges_bruto)} arestas encontradas.")
    print("Arquivo salvo: 'arestas_brutas_temporal.csv'")
else:
    print("Nenhuma coordenaÃ§Ã£o encontrada. Verifique seus parÃ¢metros.")
```

```python
# ==============================================================================
# BLOCO 3: AGREGAÃ‡ÃƒO DA REDE (Mensagens -> UsuÃ¡rios)
# ==============================================================================
print(">>> [BLOCO 3] Agregando Rede de UsuÃ¡rios...")

if not df_edges_bruto.empty:
    # 1. Identificar qual coluna tem o ID do usuÃ¡rio
    # Se vocÃª tiver 'author_id' ou 'phone', use aqui. 
    # Se for anÃ´nimo, use 'id_member_anonymous' (comum no seus scripts anteriores)
    col_user = 'id_member_anonymous' 
    
    # VerificaÃ§Ã£o de seguranÃ§a caso a coluna mude de nome
    if col_user not in df.columns:
        print(f"Aviso: Coluna '{col_user}' nÃ£o achada. Usando Ã­ndices numÃ©ricos.")
        user_map = {idx: idx for idx in range(N)}
    else:
        user_map = df[col_user].to_dict()
    
    # 2. Mapear Ã­ndices da mensagem (0, 1, 2...) para UsuÃ¡rios Reais
    df_edges_bruto['user_source'] = df_edges_bruto['source'].map(user_map)
    df_edges_bruto['user_target'] = df_edges_bruto['target'].map(user_map)
    
    # 3. Agrupar (Groupby)
    # Conta quantas vezes cada par de usuÃ¡rios interagiu
    rede_final = df_edges_bruto.groupby(['user_source', 'user_target']).size().reset_index(name='weight')
    
    # 4. Salvar
    rede_final.to_csv("rede_final_usuarios.csv", index=False, sep=";")
    print(f"Rede Agregada gerada com {len(rede_final)} conexÃµes entre usuÃ¡rios.")
    print("Arquivo salvo: 'rede_final_usuarios.csv'")
    
    # Mostra os top 5 pares mais conectados
    print("\nTOP 5 PARES MAIS COORDENADOS:")
    print(rede_final.sort_values('weight', ascending=False).head(5))

else:
    print("Pule este passo, pois nÃ£o hÃ¡ arestas brutas.")
```
